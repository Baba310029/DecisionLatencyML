```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/decision_data.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="decision_time",
    y="is_correct",
    alpha=0.3
)
plt.title("Decision Time vs Correctness")
plt.show()
```

```python
plt.figure()
sns.histplot(df["decision_time"], bins=30)
plt.title("Distribution of Decision Time")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df[[
        "decision_time",
        "task_difficulty",
        "risk_level",
        "reward_value",
        "time_pressure",
        "is_correct"
    ]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/user_clusters.csv")
```

```python
plt.figure()
sns.countplot(data=df, x="decision_style_label")
plt.title("Distribution of Decision-Making Styles")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="avg_decision_time",
    y="accuracy_rate",
    hue="decision_style_label",
    s=100
)
plt.title("Decision Speed vs Accuracy by Style")
plt.show()
```
